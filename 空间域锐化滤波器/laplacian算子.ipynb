### 代码实现

```python
import cv2
import numpy as np
img = cv2.imread("../pic/lena.png", 0)
laplacian = cv2.Laplacian(img, cv2.CV_64F)
laplacian = cv2.convertScaleAbs(laplacian) # 转为uint8
cv2.imshow("original", img)
cv2.imshow("laplacian", laplacian)
cv2.waitKey()
cv2.destroyAllWindows()
```
