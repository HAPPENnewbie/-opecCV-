# sober算子

---

## sober算子理论基础

> 先简单理解，通过像素值相减来判断是不是边界

> 水平方向的梯度计算(右边像素值减去左边像素值来判断边界)

> 垂直方向的梯度计算

> 整体算梯度计算(平方不好算,简化为右边的式子)

---

## sober算子函数以及使用

### 函数

> 该函数的一些参数

> 处理起来往往会出现相减得到负数的情况，所以不用-1，用CV_64f来扩大像素值范围

> 出现负数了怎么办

>   dx, dy

> 计算sobel梯度（推荐方式二，方式一可能检测不到边）

> 权值可以用下列函数

### 代码

```python
import cv2
import numpy as np
img = cv2.imread('../pic/lena.png',0)
sobelX = cv2.Sobel(img, cv2.CV_64F, 1, 0)  #这里是x方向的梯度
sobelY = cv2.Sobel(img, cv2.CV_64F, 0, 1)  #这里是Y方向的梯度
sobel11 = cv2.Sobel(img, cv2.CV_64F,1, 1)  # 方式一， 结果不理想
sobelX = cv2.convertScaleAbs(sobelX)  #转回uint8
sobelY = cv2.convertScaleAbs(sobelY)  #转回uint8
sobel11 = cv2.convertScaleAbs(sobel11)
sobelXY = cv2.addWeighted(sobelX, 0.5, sobelY, 0.5, 0)
sobelXY = cv2.addWeighted(sobelX, 0.5, sobelY, 0.5, 0)
cv2.imshow("original", img)
cv2.imshow("x", sobelX)
cv2.imshow("y", sobelY)
cv2.imshow("xy", sobelXY)
cv2.imshow("11", sobel11)
cv2.waitKey()
```
