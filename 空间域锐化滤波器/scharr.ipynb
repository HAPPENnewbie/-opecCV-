# scharr算子

## 理论

> scharr算子一般比sobel更加精准

## 代码

### 所用函数

> 注意：dx = 1， dy = 1会报错

> sobel算法也可以来算scharr

### 代码实现

```python
import cv2
import numpy as np
img = cv2.imread('../pic/lena.png',0)
scharrX = cv2.Scharr(img, cv2.CV_64F, 1, 0)  #这里是x方向的梯度
scharrY = cv2.Scharr(img, cv2.CV_64F, 0, 1)  #这里是Y方向的梯度

scharrX = cv2.convertScaleAbs(scharrX)  #转回uint8
scharrY = cv2.convertScaleAbs(scharrY)  #转回uint8

scharrXY = cv2.addWeighted(scharrX, 0.5, scharrY, 0.5, 0)

cv2.imshow("original", img)
cv2.imshow("x", scharrX)
cv2.imshow("y", scharrY)
cv2.imshow("xy", scharrXY)
cv2.waitKey()
```
