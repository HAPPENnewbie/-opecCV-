```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show(img):
    if img.ndim == 2:
        plt.imshow(img, cmap='gray')
    if img.ndim == 3:
        plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
    plt.show()
```

```python
img = cv.imread('../pic/1.jpg')
show(img)
```

## 

```python
M = np.array([
    [1, 0, 100],
    [0, 1, 50]
], dtype = np.float32)
```

```python
img2 = cv.warpAffine(img, M, (600, 400))
show(img2)
```

## 

```python
# 
M = np.array([
    [1, 0.2, 0],
    [0, 1, 0]
], dtype = np.float32)
```

```python
img3 = cv.warpAffine(img, M, (600, 400))
show(img3)
```

```python
# 
M = np.array([
    [1, 0, 0],
    [0.2, 1, 0]
], dtype = np.float32)
```

```python
img4 = cv.warpAffine(img, M, (600, 400))
show(img4)
```

## 

```python
# 
M = np.array([
    [-1, 0, img.shape[1]],
    [0, 1, 0]
], dtype = np.float32)
```

```python
img5 = cv.warpAffine(img, M, (img.shape[1], img.shape[0]))
show(img5)
```

```python
# 
M = np.array([
    [1, 0, 0],
    [0, -1, img.shape[0]]
], dtype = np.float32)
```

```python
img6 = cv.warpAffine(img, M, (img.shape[1], img.shape[0]))
show(img6)
```

```python
# opencv
img6 = cv.flip(img, 1)
img7 = cv.flip(img, 0)
img8 = cv.flip(img, -1)
show(np.hstack([img, img6, img7, img8]))
```
