# 图像的放大缩小

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show(img):
    if img.ndim == 2:
        plt.imshow(img, cmap='gray')
    if img.ndim == 3:
        plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
    plt.show()
```

```python
img = cv.imread('../pic/1.jpg')
show(img)
```

```python
# 图片裁剪
dog = img[100:250, 50:150, :]    # 高从100 到250， 宽从50 到150 ， 第三个参数是选择通道，":" 表示全选
show(dog)
```

```python
# 图片放大  放大的插值法一般用下面这个
#opencv x， y
img2 = cv.resize(img, (400, 500), interpolation=cv.INTER_LINEAR)  
show(img2)
```

```python
# 图片缩小  缩小的插值法一般用下面这个
#opencv x， y
img2 = cv.resize(img, (100, 200), interpolation=cv.INTER_AREA)  
show(img2)
```
