# 简单绘图

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 折线图

```python
# 建立函数
x = np.arange(2, 20)
y = 2 * x + np.random.randint(5, 20, 18)  
# 构图
plt.plot(x, y, '*-', color='r')
plt.show()
```

```python
x = np.linspace(0, 1, 100)
y1 = np.power(x, 0.5)
y2 = x
y3 = np.power(x, 1.5)

plt.plot(x, y1, label='0.5')
plt.plot(x, y2, label='1.0')
plt.plot(x, y3, label='1.5')
plt.legend()       # 加上这个才显示label
plt.xlabel('r')
plt.ylabel('s')       
plt.grid()            #表格背景
plt.xlim([0, 1])      #x轴范围
plt.ylim([0, 1])
plt.show()
```

## 直方图

```python
a = np.random.randint(0, 101, 1000)
```

```python
a
```

```python
plt.hist(a, rwidth=0.9, color='g')
plt.show()
```

```python
bins = np.arange(-0.5, 101, 1)  # 直方图的柱数，即要分的组数，默认为10
```

```python
plt.hist(a, bins, rwidth=0.8)
plt.show()
```
