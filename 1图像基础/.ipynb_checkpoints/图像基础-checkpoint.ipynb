# 图像基础

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
# 自定义一个函数来输出
def show(img):
    if img.ndim == 2:  
        plt.imshow(img, cmap='gray', vmin=0, vmax=1)    #在显示灰度图像时，要设定上vmin=0以及vmax=255。否则imshow函数会自动进行数值调整。
    else:
        plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
    plt.show()
```

```python
# 随机建立一个二维矩阵，  灰度图
A = np.random.randint(0, 256, (2, 4), dtype=np.uint8)
A
```

```python
show(A)
```

```python
# 随机建立一个二维矩阵，  彩色图
B = np.random.randint(0, 256, (2, 4, 3), dtype=np.uint8)
B
```

```python
show(B)
```

```python
# 8位整型图像   这里的-1自动变为255了，可以类比时间的进制
np.uint8([-1, 0, 10, 255, 300])
```

```python
# 浮点数图像
A2 = np.float32(A)
A2 /= 255   # 0 ~ 1 之间来表示灰度
A2
```

```python
show(A2)  # 这里注意前面自定义show函数要修改vim=0和vmax=1才可以正常显示
```
