# 通道的分离与合并

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show(img):
    if img.ndim == 2:
        plt.imshow(img, cmap='gray')
    if img.ndim == 3:
        plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
    plt.show()
```

```python
img = cv.imread('pic/1.jpg')
show(img)
```

```python
# 分离
b, g, r = cv.split(img)
show(r)
```

```python
# 合并
img2 = cv.merge([b, g, r])
show(img2)
```

# 彩色图转换为灰度图

```python
gray1 = 1/3*b + 1/3*g + 1/3*r
show(gray1)
```

```python
gray1 # 可以看到类型是浮点
```

```python
# 转化为uint8类型
gray2 = np.uint8(gray1)
gray2
```

```python
# 另一种彩色图转灰度图， 利用opencv里的函数
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)          # cvt = convert  2 = to
show(img_gray)
```

# 图像二值化

```python
thresh = 125 # 
img_gray[img_gray > thresh] = 255
img_gray[img_gray <= thresh] = 0
show(img_gray)
```

```python
# 利用opencv接口来图像二值化
ignore, img_bin = cv.threshold(gray1, 125, 255, cv.THRESH_BINARY)  # 大于125的变为255， 反之等于0
show(img_bin)
```
