# 图片读取显示

## 导入库

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

## 图片导入与显示

```python
#导入图片方法一
img = cv.imread('pic/1.jpg')
```

```python
# 将img保存到test中
cv.imwrite('test/1.jpg', img)
```

```python
# 显示图片
cv.imshow('1',img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# 导入图片 方法二
plt.imshow(img)
plt.show()   
```

```python
# 方法二回归常见色彩  RGB BGR
img2 = cv.cvtColor(img, cv.COLOR_BGR2RGB)  # 转换一下颜色通道
plt.imshow(img2)
plt.show()
```

```python
# 图片以灰度图来显示  方法一
img3 = cv.imread('pic/1.jpg', 0)
cv.imshow('1', img3)
cv.waitKey(0)
cv.destroyAllWindows()

# 图片以灰度图来显示  方法二
plt.imshow(img3, cmap='gray')
plt.show()
```

```python
# 彩图维度是3维， 灰度图维度是2维
img.ndim, img3.ndim
```

```python
# 可以通过这个来判断一个图片是不是灰度图
def show(img):
    if img.ndim == 2:
        plt.imshow(img, cmap='gray')
    if img.ndim == 3:
        plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
    plt.show()
```

```python
#测试刚写的函数
show(img)
```
