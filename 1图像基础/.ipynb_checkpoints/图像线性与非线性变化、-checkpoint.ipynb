# 图像线性变化与非线性变化

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
def show(img):
    if img.ndim == 2:
        plt.imshow(img, cmap='gray')
    if img.ndim == 3:
        plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
    plt.show()
```

```python
img = cv.imread('pic/1.jpg', 0)
show(img)
```

```python
# 线性变化    
b = 20
k = 2
img2 = b + k * img.astype(np.int32) # 将img转换为int32类型
img2 = np.clip(img2, 0, 255)   # 小于0的变为0， 大于255的变为255
show(img2)       # 可以看到亮度变化
```

```python
img2.max()
```

```python
# 使用opencv内置函数来实现线性变化
img3 = cv.convertScaleAbs(img, alpha = 2, beta = 20)  # alpha 相当于a   beta相当于b
show(img3)
```

```python
# 非线性变换
img4 = 10 + np.log(img.astype(np.float32) + 1) / 0.1
show(img4)
```

```python
# 伽马变化 
img01 = img / 255   
img05 = np.power(img01, 0.5) * 255  # 指数为0.5  ， 结果变亮
img15 = np.power(img01, 1.5) * 255  # 指数为1.5  ， 结果变暗
show(np.hstack([img05, img, img15]))
```
