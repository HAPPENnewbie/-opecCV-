### 对图像卷积

> 一般用小的来对大的来卷，小的称为“卷积核”（算子，滤波器）。
> 依次变换大矩阵的中心来与卷积核来进行卷积运算

> 我们看到卷积过后图像变小了，为了保持不变，我们在卷积后要进行填充为何原图像大小一致。填充规则有很多，比如复制填充（见下图）

## 代码

### 公式小览

> openCV的卷积不是真的卷积，它是正着对应乘的，如果要真的卷积，需要flip一下

### 对卷积进行计算

```python
import numpy as np
import cv2 as cv
```

```python
img = np.ones((5,5))
kernel = np.ones((3,3))

print(img)
print(kernel)
```

```python
# 进行卷积
img2 = cv.filter2D(img, -1, kernel)

print(img2)
```
