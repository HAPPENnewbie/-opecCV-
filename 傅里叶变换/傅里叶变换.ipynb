## 傅里叶变换原理

> 以做饮料来理解，一个是用时间（空间）域来表示，另一个是从频率的角度来表示。可以看出频率角度明显比时空域要简洁

> 注意二者横坐标

> 但是会有更加精细的时间。比如例子中的都是18开始但是秒不同，这时候就需要用**相位**来描述。<br>
> 相位是与时间差有关的一个描述

> 傅里叶得出：**任何连续的周期信号，可以由一系列正弦曲线得出**

> 傅里叶变换可以使得空间域和频域相互转换（两者是可逆的）

> 不是同时开始的一组余弦函数，叠加时需要用相位来体现开始的时间

## numpy实现逆傅里叶变换

> 频谱恢复到原始图像成为逆变换

### 函数介绍

> 得到一个复数的数组

### 代码

```python
# 导入库
import cv2
import numpy as np
import matplotlib.pyplot as plt
# 读取图像
img = cv2.imread('../pic/lena.png', 0)  # 0 表示读入的灰度图
# 进行傅里叶变换
img2 = np.fft.fft2(img)
# 移动低频到中心位置
fshift = np.fft.fftshift(img2)
# 将复数调整为为0-255可以显示的频谱
result = 20*np.log(np.abs(fshift))
# 显示原图
plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.title('original')
plt.axis('off')
# 显示频谱
plt.subplot(122)
plt.imshow(result, cmap='gray')
plt.title('result')
plt.axis('off')
# 显示
plt.show()
```

> 注意  低频对应细节，高频对应边界

## numpy实现逆傅里叶变换

### 函数介绍

### 代码实现

```python
# 注意
# 导入库
import cv2
import numpy as np
import matplotlib.pyplot as plt
# 读取图像
img = cv2.imread('../pic/lena.png', 0)  # 0 表示读入的灰度图
# 进行傅里叶变换
f = np.fft.fft2(img)
# 低频移到中心
fshift = np.fft.fftshift(f)
# 低频移动回去
ishift = np.fft.ifftshift(fshift)
# 逆傅里叶变换
iimg = np.fft.ifft2(ishift)
# 将复数转换为实数得以显示
iimg = np.abs(iimg)
#显示原始图像
plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.title('original')
plt.axis('off')
#显示目标图像
plt.subplot(122)
plt.imshow(iimg, cmap='gray')
plt.title('iimg')
plt.axis('off')
plt.show()  # 显示起来两者一样
```
